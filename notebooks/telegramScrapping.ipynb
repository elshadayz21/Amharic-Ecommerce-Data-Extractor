```python
from amseg import AmharicSegmenter #used for Amharic text segmentation, segmentation is the process of dividing text into sentences and words.
sent_punct = []
word_punct = []
segmenter = AmharicSegmenter(sent_punct, word_punct) # Create an instance of AmharicSegmenter with custom punctuation lists


#we need to install pytesseract for text extraction from images, which is required by amseg for OCR functionality
```

```python
# Function to tokenize Amharic text
def tokenize_amharic_text(text):
    """
    Tokenizes Amharic text using nltk word_tokenize 
    """
    tokens = segmenter.amharic_tokenizer(text) #tokenizing means breaking down the text into smaller units, such as words or phrases.
   
    return tokens
```

```python
def normalize_amharic_text(text):
    """
    Normalizes Amharic text by removing non-Amharic characters and handling case sensitivity.
    """
    #normalization is the process of converting text to a standard format, such as lowercasing, removing punctuation, etc.
    text = text.lower()
    # Replace special punctuation marks with a space
    text = re.sub(r'[፡።፣፤፥፦]', ' ', text) # Replace Amharic punctuation with space, r is a raw string literal
    # Remove any extra spaces
    text = re.sub(r'\s+', ' ', text).strip() # Remove extra spaces and strip leading/trailing spaces, strip() removes leading and trailing whitespace
    return text
```

```python
def preprocess_amharic_text(text):
    normalized_text = normalize_amharic_text(text)
    tokens = tokenize_amharic_text(normalized_text)
    print("Tokens:", tokens)
    print("Normalized Text:", normalized_text)
    return tokens
```

```python
data["Product Tokens"] = data["Product Description"].apply(preprocess_amharic_text)
data.to_csv("telegram_data_preprocessed.csv", index=False) # Save the preprocessed data to a new CSV file
```

```python
product_keywords = ['የልብስ','ጫማ','ጥፍር','መገጣጠሚያ','ጥሬ ንግድ','ምግብ','መጠጥ','ማስቀመጫ','ልብስ','ብሩሽ','ቁምሳጥን',
                    'የሞተ','ቆዳን','ቆሻሻን','ለማፅዳት','ከብረት','የተሰራ','የተለያዩ','ዕቃወች','እንቁላል','መያዝ','የሚችል','የቦርጭ',
                    'ስብን','ለማቅለጥ','መኪና','ለማጠብ','ግቢ','ለማፅዳት','መቀቀል','የሚያስችል','ጸጉር','የሚያስውብ','ልጆች','የሚማሩበት','ለፅሁፍ','ለስዕል','መከላከያ',
                     'መዓዛን','የሚሰጥ','በመንሸራተት','ከሚፈጠር','አደጋ','የሚከላከ','ከወፍራም','ብረት','ጁስ','ለማዘጋጀት','ለዘይት','መቀነሻ','ልዮ',
                    'መአዛን','የሚያላብስ','ለእግር','ልስላሴ','ስጋ','ጨጓራ','ሽንኩርት','ድቅቅ','የሚያደርግ','ከአልሙኒየም']
location_keywords = ['ቢሮ','ፒያሳ','ጊዮርጊስ','ቦታ','አድራሻ','መገናኛ','ሞል','ቁ','1','2','መሰረት','ደፋር','ሁለተኛ','ፎቅ',
                     '05','06','ራመት_ታቦር_ኦዳ_ህንፃ','1ኛ','አደባባይ','ሱቅ']
price_keywords = ['ብር', 'ዋጋ']
```

```python

def label_tokens(tokens):
    labels = []
    inside_price = False # This variable will track if we are inside a price context
    inside_loc=False # This variable will track if we are inside a location context
    inside_prod = False # This variable will track if we are inside a product context
    for token in tokens:
        if token in product_keywords and inside_prod: # If the token is a product keyword and we are already inside a product context
            labels.append('I-Product') #if inside-prod is True, we append 'I-Product' to the labels list
        elif token in product_keywords:
            labels.append('B-Product')
            inside_prod =True    
        elif token in location_keywords and inside_loc:
            labels.append('I-LOC')   # Continuing a location name
        elif token in location_keywords:
            labels.append('B-LOC') # Beginning of a location
            inside_loc = True
        elif token in price_keywords: 
            labels.append("B-PRICE")    # Start of a price expression (e.g., '$')
            inside_price = True
        elif token.isdigit() and inside_price:  
            labels.append("I-PRICE")    # Continuation of price (e.g., '100')
        elif token.isdigit(): 
            labels.append("O")  
            inside_price = False
        else:
            labels.append("O")   # Number not part of a price
            inside_price = False
        
    return labels
```

```python
tokens1 = data['Product Tokens']
```

```python
labeled_tokens = data['Product Tokens'].apply(label_tokens)
print(labeled_tokens.head(10))
```

```python
coNLL_data=[] # Initialize an empty list to store the CoNLL formatted data, coNLL format is a common format for representing tokenized text with labels, often used in NLP tasks like Named Entity Recognition (NER).
for token, label in zip(tokens1, labeled_tokens): #zip combines the tokens and labels into pairs, allowing us to iterate over both simultaneously.
    for t,l in zip(token,label):   
        coNLL_data.append(f"{t} {l}")
    coNLL_data.append("")
coNLL_data="\n".join(coNLL_data)
    
```

```python
# Save to a text file in CoNLL format
with open('labeled_data.conll', 'w', encoding='utf-8') as f: #conll format is a common format for representing tokenized text with labels, often used in NLP tasks like Named Entity Recognition (NER).
    f.write(coNLL_data)

print("Labeled data has been saved in CoNLL format.")
```
